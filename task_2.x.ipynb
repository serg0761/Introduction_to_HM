### Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. 

```python
def lenght_of_vector(x1, x2, y1, y2):
    return np.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
```

### Напишите код на Python, реализующий построение графиков:
 окружности,
 эллипса,
 гиперболы.


```python
def half_of_circle(x, x0, y0, R):
    return np.sqrt(R ** 2 - (x - x0) ** 2) + y0
```

```python
x = np.arange(-100, 100, 0.0001)
y1 = half_of_circle(x, 0, 0, 3)
y2 = -half_of_circle(x, 0, 0, 3)
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.title('Окружность')
```

```python
def half_of_ellipse(a, b, x):
    return np.sqrt(b * (1 - x ** 2 / a ** 2))
```

```python
y1 = half_of_ellipse(2, 1, x)
y2 = -half_of_ellipse(2, 1, x)
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.title('Эллипс')
```

```python
def half_of_hyperbole(a, b, x):
    return np.sqrt(b * (x ** 2 / a ** 2 - 1))
```

```python
x = np.arange(-5, 5, 0.001)
y1 = half_of_hyperbole(1, 2, x)
y2 = - half_of_hyperbole(1, 2, x)
plt.plot(x, y1)
plt.plot(x, y2)
plt.title('Гипербола')
```

## Нарисуйте трехмерный график двух параллельных плоскостей

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = 2 * X + 5 * Y
Z1 = 2 * X + 5 * Y + 25
ax.plot_wireframe(X, Y, Z)
ax.plot_wireframe(X, Y, Z1, color='r')
ax.scatter(0, 0, 0, 'z', 50, 'red')
show()
```

## Нарисуйте трехмерный график двух любых поверхностей второго порядка.


```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-10, 10, 0.5)
Y = np.arange(-10, 10, 0.5)
X, Y = np.meshgrid(X, Y)
Z = X ** 2 +  Y ** 2 + 10
Z1 =  X ** 2 -  Y ** 2 - 15
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z1, color='r')
ax.scatter(0, 0, 0, 'z', 50, 'red')
show()
```
