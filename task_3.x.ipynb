Докажите, что при ортогональном преобразовании сохраняется расстояние между точками.


```python
def distance(x1, y1, x2, y2):
    return np.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2)
```

```python
x1 = 2
y1 = 5
x2 = 10
y2 = 3
```

```python
base_distance = distance(x1, y1, x2, y2)
```

```python
base_distance
```

```python
a = 3
b = 5
alpha = np.pi / 3
```

```python
x2_1 = (x1 - a) * np.cos(alpha) + (y1 - b) * np.sin(alpha)
y2_1 = -(x1 - a) * np.sin(alpha) + (y1 - b) * np.cos(alpha)
x2_2 = (x2 - a) * np.cos(alpha) + (y2 - b) * np.sin(alpha)
y2_2 = -(x2 - a) * np.sin(alpha) + (y2 - b) * np.cos(alpha)
```

```python
new_distance = distance(x2_1, y2_1, x2_2, y2_2)
base_distance == new_distance
```

Напишите код, который будет переводить полярные координаты в декартовы.


```python
def transform_to_descartes(r, phi):
    x = r * np.cos(phi)
    y = r * np.sin(phi)
    return x, y
```
