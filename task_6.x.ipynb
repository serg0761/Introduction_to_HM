Найдите нормальное псевдорешение недоопределенной системы:


```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
A = np.array([[1, 2, -1], [8, -5, 2]])
b = np.array([1, 12])
```

```python
def Q(x, y, z):
    return (x **2 + y ** 2 + z ** 2)
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-10, 10, 0.5)
Y = np.arange(-10, 10, 0.5)
X, Y = np.meshgrid(X, Y)
ax.plot_surface(X, Y, Q(X, Y, -1 - X - 2 * Y))
ax.plot_surface(X, Y, Q(X, Y, 6 - 4*X + 5 / 2 * Y), color='r')
ax.scatter(0, 0, 0, 'z', 50, 'red')
show()
```
